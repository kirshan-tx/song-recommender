# Missing Features

***

The following code was used to obtain the audio features of the songs that were missing from favFeatures; Using [Spotipy](https://spotipy.readthedocs.io) - a lightweight python library for the Spotify Web API.

## Data

- fav is a dataframe of my favourite songs, which were songs that I've listened to more than or equal to 9 times.
- favFeatures is a dataframe of audio features for the corresponding track in fav dataframe.

```python
#import libraries
import pandas as pd 
import spotipy 
import numpy as np

#spotipy authorization
sp = spotipy.Spotify() 
from spotipy.oauth2 import SpotifyClientCredentials 

cid="CLIENT_ID" 
secret="CLIENT_SECRET"
username="USERNAME"

client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret) 
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager) 

sp.trace=False 
```

```python
#load in fav and favFeatures dataset
fav = pd.read_csv('fav.csv')
favFeatures = pd.read_csv('favFeatures.csv')
```

#### Of the 191 tracks in fav, only 68 corresponding audio features are present in favFeatures.  

```python
#unqiue tracks in fav 
fav_track1 = fav.track.unique()
fav_track1[:15] #first 15 tracks
```

```python
#unqiue tracks in favFeatures  
fav_track2 = favFeatures.track.unique()
fav_track2[:15] #first 15 tracks
```

```python
missingTracks = dict()

for track in fav_track1:
    if track not in fav_track2:
        #using spotipy to search track name to get track id
        track_id = sp.search(q=' track:' + track, type='track')
        try:
            #populate missingTraks with each track name: track id
            missingTracks[track] = track_id['tracks']['items'][0]['id']
        except IndexError:
            print("Unable to find id for: " + track)
```

```python
#make missingTracks dictionary into dataframe called ids
ids = pd.DataFrame(missingTracks.items(), columns=['track', 'id'])
ids.head()
```

Since there is only 3 songs, I manually found their respective track ids.

- Don't Check On Me (feat. Justin Bieber & Ink) : 7ILEWkpfdK6AjH2D7jppWk

```python
track1 = pd.Series(data={"track":"Honesty (Remix) [feat. Jessie Reyez]", "id":"47r6MeqPoeQgCsa5vOQxoo"})
```

```python
track2 = pd.Series(data={"track":"Don't Check On Me (feat. Justin Bieber & Ink)", "id":"7ILEWkpfdK6AjH2D7jppWk"})
```

```python
track3 = pd.Series(data={"track":"I'm the One (feat. Justin Bieber, Quavo, Chance the Rapper & Lil Wayne)", "id":"3DXncPQOG4VBw3QHh3S817"})
```

```python
#append the 3 tracks
ids = ids.append(track1, ignore_index=True)
ids = ids.append(track2, ignore_index=True)
ids = ids.append(track3, ignore_index=True)
ids.tail()
```

```python
def get_features(df):
    playlist_features_list = ["id", "track", "duration_ms", "danceability","energy","key","loudness","mode", "speechiness","acousticness", "instrumentalness","liveness","valence","tempo", "time_signature"]
    playlist_df = pd.DataFrame(columns = playlist_features_list)
    
    for i in range(len(df)): 
        playlist_features = {} 
        playlist_features['id'] = df['id'][i]
        playlist_features['track'] = df['track'][i]
        playlist_features['favourite'] = np.int32(1)
        
        audio_features = sp.audio_features(playlist_features['id'])[0]

        for feature in playlist_features_list[2:]:
            playlist_features[feature] = audio_features[feature]

        
        track_df = pd.DataFrame(playlist_features, index = [0])
        playlist_df = pd.concat([playlist_df, track_df], ignore_index = True)
    
    return playlist_df
```

```python
#dataframe with audio features of songs that were missing from favFeatures
missingFeatures = get_features(ids)
missingFeatures
```

```python
#save to csv
missingFeatures.to_csv('missingFeatures.csv', index=False)
```
